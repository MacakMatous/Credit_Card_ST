```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Bank_Churners.csv")
```

```python
# Create a scatter plot for age vs. credit limit
plt.scatter(df['Customer_Age'], df['Credit_Limit'], alpha=0.5)
plt.xlabel("Customer Age")
plt.ylabel("Credit Limit")
plt.plot()
```

```python
# Create a bar chart for education levels
edu_counts = df['Education_Level'].value_counts()
plt.bar(edu_counts.index, edu_counts.values)
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.plot()
```

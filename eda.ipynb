## Feature explaination

| **Feature Name** | **Explaination** |
| --- | --- |
| ID | Identification number in form of a sequence |                      
| CLIENTUM | Internal client number | 
| Attrition_Flag | Whether or not is the customer actively existing and using our services. (Existing Customer, Attrited Customer) |                
| Customer_Age | Age of the customer|     
| Gender | Gender of the customer|         
| Dependent_count | Count of dependents (family members) associated with the customer.|
| Education_Level | Highest achived level of education of the customer |
| Marital_Status | Marital Status of the customer| 
| Income_Category | Income level of the customer (Less than $40K, $40K - $60K, $60K - $80K, $80K - $120K, $120K +, Unknown)|
| Card_Category | Category of the clients card (Blue, Silver, Gold,Platinum)|         
| Months_on_book | Duration (in months) that a customer has held a credit card.|          
| Total_Relationship_Count | Total number of relationships (accounts, services, products) a customer has with the bank.|
| Months_Inactive_12_mon | Number of inactive months (not using credit card or other services) in the last 12 months. |  
| Contacts_Count_12_mon | Number of times the customer contacted the bank in the last 12 months. |   
| Credit_Limit | Maximum amount a customer can borrow on their credit card. |            
| Total_Revolving_Bal | Total revolving balance (outstanding debt carried over month to month).|     
| Avg_Open_To_Buy | Average open-to-buy amount (credit limit minus current balance).|         
| Total_Amt_Chng_Q4_Q1 | Total change in spending between Q4 and Q1.|    
| Total_Trans_Amt | Sum of all transactions made by the customer during a specific period. |         
| Total_Trans_Ct | Total number of transactions conducted by the customer. |          
| Total_Ct_Chng_Q4_Q1 | Total count of changes in transactions between Q4 and Q1.|     
| Avg_Utilization_Ratio | Average utilization ratio (percentage of credit used relative to the limit).|   

## Definition of KPIs

**Customer attrition (churn) rate:** <br>
This metric represents the percentage of customers who have stopped using our product or service during a specific time period. Retaining existing customers is generally more cost-effective than acquiring new ones.

**Transaction Volume (Total_Trans_Ct):** <br>
This KPI measures the total number of credit card transactions processed within a given period.<br>
It reflects the level of demand for our products or services and helps assess our business growth.

**Average Ticket Size (Avg_Open_To_Buy):** <br>
The average amount spent per transaction is a vital indicator to gauge the purchasing behavior of our <br> customers, identify trends, and evaluate revenue potential.

## EDA graphs

```python
# Count the occurrences of each category
category_counts = df['Card_Category'].value_counts().sort_values(ascending=False)

# Set a modern style
plt.style.use('seaborn-v0_8-darkgrid')

# Create the bar plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(category_counts.index, category_counts.values, color=["Lightblue", "Silver", "Gold", "Silver"], edgecolor='black')

# Add the counts above the histogram bars
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{int(height)}', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=12, color='black')

# Add labels and title
ax.set_xlabel('Card Category', fontsize=14, color='black')
ax.set_ylabel('Frequency', fontsize=14, color='black')
ax.set_title('Histogram of Card Category (Sorted by Frequency)', fontsize=16, color='black', weight='bold')

# Customize tick parameters
ax.tick_params(axis='x', colors='black', labelsize=12)
ax.tick_params(axis='y', colors='black', labelsize=12)

# Customize the background color
fig.patch.set_facecolor('#f5f5f5')
ax.set_facecolor('#f5f5f5')

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Create a count plot
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Education_Level', data=df, hue='Education_Level', palette='viridis', order=df['Education_Level'].value_counts().index, legend=False)


# Add labels and title
ax.set_xlabel('Education Level', fontsize=14, color='black')
ax.set_ylabel('Frequency', fontsize=14, color='black')
ax.set_title('Distribution of Education Levels', fontsize=16, color='black', weight='bold')

# Customize the tick parameters
ax.tick_params(axis='x', colors='black', labelsize=12, rotation=45)
ax.tick_params(axis='y', colors='black', labelsize=12)

# Add counts above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

# Customize the background color
plt.gcf().set_facecolor('#f5f5f5')
ax.set_facecolor('#f5f5f5')

# Show plot
plt.tight_layout()
plt.show()
```
